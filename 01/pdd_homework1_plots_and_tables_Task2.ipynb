```python
#-----------------------------------------------------------------------------------------------------#
#Use tools available for Python to plot charts and tables to visualize/display this information:      #
#   -Find the departments/study programmes/supervisors/reviewers with highest numbers of thesis       #
#     and come up with some nice plots and tables to depict their numbers.                            #
#-----------------------------------------------------------------------------------------------------#

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # standard convention for importing the plotting tool
import matplotlib
%matplotlib inline
matplotlib.style.use('ggplot')

#--------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------#
#--------FUNCTIONS--------FUNCTIONS--------FUNCTIONS--------FUNCTIONS--------FUNCTIONS--------FUNCTIONS--------#
#--------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------#

def drawPlotAndStoreMaxValue(category_name, plot_color, y_axis_graduation) :
    #we define it outside function
    global max_values
    
    print(category_name)
    
    theses_df = all_theses_df[category_name]
    theses = []

    for thesis in theses_df :
        if thesis in theses :
            continue
        else :
            theses.append(thesis)

    df = pd.DataFrame(columns = [category_name,'NUMBER_OF_THESES'])

    numOfThesesPerCategory = []

    for thesis in theses :
            
                temp_df = all_theses_df[(all_theses_df[category_name]==thesis)]
                count = len(temp_df)

                if count>0 :
                    tmp_dict = {
                        "name" : thesis,
                        "num_of_th" : count
                    }
                    numOfThesesPerCategory.append(tmp_dict)
                    df = df.append({category_name : thesis, 'NUMBER_OF_THESES' : count}, ignore_index = True)

    sort_df = df.sort_values(by=['NUMBER_OF_THESES'], ascending=False)

    tmp_cat = category_name
    tmp_name = sort_df[category_name].iloc[0]
    tmp_num = sort_df['NUMBER_OF_THESES'].iloc[0]

    tmp_dict = {
        "cat" : tmp_cat,
        "name" : tmp_name,
        "num" : tmp_num
    }

    max_values.append(tmp_dict)

    #in one plot there is just 20 maximum values
    sort_df = sort_df[:20]

    plt.figure(figsize=(12, 10), dpi= 80, facecolor='w', edgecolor='k')
    plt.yticks(np.arange(0, 6000, y_axis_graduation))
    plt.xticks(rotation=90)
    plt.bar(sort_df[category_name], sort_df['NUMBER_OF_THESES'], color=plot_color)
    plt.show()
    
def createTableWithMaxValues():
    global max_values
    
    df = pd.DataFrame(columns = ['CATEGORY','NAME','NUMBER_OF_THESES'])
    for value in max_values :
        tmp_cat = value.get("cat")
        tmp_name = value.get("name")
        tmp_num = value.get("num")
        
        df = df.append({'CATEGORY' : tmp_cat, 'NAME' : tmp_name, 'NUMBER_OF_THESES' : tmp_num}, ignore_index = True)
        
    display(df)
    
#--------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------#
#-----END OF FUNCTIONS-----END OF FUNCTIONS-----END OF FUNCTIONS-----END OF FUNCTIONS-----END OF FUNCTIONS-----#
#--------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------------------#  
  
#prepare all data
all_theses_df = pd.read_csv('pdd_homework1_wz_all_data.csv', sep='|')
all_theses_df['DATE_OF_DEFENCE'] = all_theses_df['DATE_OF_DEFENCE'].str[:4]
del all_theses_df['Unnamed: 0']

#to store maximum value of each cathegory to put it later into one table to compare
max_values = []

#just uncomment this section below to draw plots and create table

#drawPlotAndStoreMaxValue("STUDY_PROGRAMME", "Blue",200)
#drawPlotAndStoreMaxValue("DEPARTMENT_NAME", "Red",100)
#drawPlotAndStoreMaxValue("SUPERVISOR_NAME", "Yellow",10)
#drawPlotAndStoreMaxValue("REVIEWER_NAME", "Black",5)
#createTableWithMaxValues()
```
