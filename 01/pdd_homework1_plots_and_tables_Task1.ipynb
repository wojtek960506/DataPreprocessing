```python
#--------------------------------------------------------------------------------------------------#
#Use tools available for Python to plot charts and tables to visualize/display this information:   #
#    Number of defended theses per year for CTU/Faculties. Distinguish the type of thesis.         #
#--------------------------------------------------------------------------------------------------#


import pandas as pd

#get theses info from csv file
all_theses_df = pd.read_csv('pdd_homework1_wz_all_data.csv', sep='|')
all_theses_df['DATE_OF_DEFENCE'] = all_theses_df['DATE_OF_DEFENCE'].str[:4]
del all_theses_df['Unnamed: 0']

#get all names of faculties
fac_name_df = all_theses_df['FACULTY_NAME']
fac_names = []

for f in fac_name_df :
    if f in fac_names :
        continue
    else :
        fac_names.append(f)

#get all years of defence       
th_dates_df = all_theses_df['DATE_OF_DEFENCE']
th_years = []

#there is try block because sometimes year is not available
for d in th_dates_df :
    try :
        temp_y = d[:4]
        if temp_y in th_years :
            continue
        else :
            th_years.append(temp_y)
    except TypeError :
        continue

#get all types of theses
th_types_df = all_theses_df['THESIS_TYPE']
th_types = []

for t in th_types_df :
    if t in th_types :
        continue
    else :
        th_types.append(t)
    
    if len(th_types)==2 :
        break

#save seeking informations into data frame ans show them
df = pd.DataFrame(columns = ['FACULTY_NAME','YEAR','THESIS_TYPE','NUMBER_OF_THESES'])
numOfThPerYearForFacs = []

for fac_name in fac_names :
    for th_year in th_years :
        for th_type in th_types :
            #we count theses of special faculty, year and type
            temp_df = all_theses_df[(all_theses_df.FACULTY_NAME==fac_name)&(all_theses_df.DATE_OF_DEFENCE==th_year)&(all_theses_df.THESIS_TYPE==th_type)]
            count = len(temp_df)
            
            #in some years in some faculties there was no theses of such type
            if count>0 :
                tmp_dict = {
                    "fac_name" : fac_name,
                    "th_year" : th_year,
                    "th_type" : th_type,
                    "num_of_th" : count
                }
                numOfThPerYearForFacs.append(tmp_dict)
                df = df.append({'FACULTY_NAME' : fac_name, 'YEAR' : th_year, 'THESIS_TYPE' : th_type, 'NUMBER_OF_THESES' : count}, ignore_index = True)

sort_df = df.sort_values(by=['FACULTY_NAME','YEAR'], ascending=True)



print (sort_df.to_string(index=False))
#display(sort_df)
```
